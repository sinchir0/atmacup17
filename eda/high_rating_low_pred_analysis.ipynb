```python
import pandas as pd

data = pd.read_csv(
    "../trained_models/e020-auxiliary-loss/valid_dataset_e020-auxiliary-loss.csv"
)
```

```python
data["Title"] = data["Title"].fillna("")
data["Review Text"] = data["Review Text"].fillna("")

data["full_text"] = data["Title"] + " " + data["Review Text"]

data["Title_word_num"] = data["Title"].apply(lambda x: len(x.split()))
data["Review_word_num"] = data["Review Text"].apply(lambda x: len(x.split()))
data["full_text_word_num"] = data["full_text"].apply(lambda x: len(x.split()))
```

```python
high_pred_0_label = data[(data["valid_pred"] > 0.7) & (data["label"] == 0)]
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.title("high_pred_0_label")
high_pred_0_label["Age"].hist()
plt.xlim(0, 100)
plt.subplot(1, 2, 2)
plt.title("all data")
data["Age"].hist()
plt.xlim(0, 100)
```

```python
# 大した差はない
```

```python
COL = "Rating"
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.title("high_pred_0_label")
high_pred_0_label[COL].value_counts(normalize=True).sort_index().plot(kind="bar")
plt.subplot(1, 2, 2)
plt.title("all data")
data[COL].value_counts(normalize=True).sort_index().plot(kind="bar")
```
